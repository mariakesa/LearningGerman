```python
path='/media/maria/DATA1/Documents/articles.txt'
```

```python
with open(path) as f:
    lines = f.readlines()
print(lines[0])
#sentences=lines.split('.')
print(len(lines))
```

```python
article_lst=[]

for line in lines:
    article_lst.append(line.split('.'))
```

```python
#Print sentence
word='anfang'

for article in article_lst:
    for sentence in article:
        words=sentence.split(' ')
        for wd in words:
            wrd=wd.lower()
            if word==wrd:
                print(sentence)
                print('*********************************************')
            
```

```python
#Print entire article

word='anfang'

for j in range(len(lines)):
    for sentence in article_lst[j]:
        words=sentence.split(' ')
        for wd in words:
            wrd=wd.lower()
            if word==wrd:
                print(lines[j])
                print('*********************************************')
```
